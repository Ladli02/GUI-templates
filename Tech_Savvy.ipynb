### MODEL SELECTION AND EVALUTION

```python
import pickle
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, BaggingRegressor, ExtraTreesRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, HuberRegressor, SGDRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

def regression_models_with_feature_selection(df, target_variable, k_best_features=None, test_size=0.2, random_states=[42]):
    models_data = {'model_name': [], 'Random State': [], 'Accuracy': [], 'MSE': []}
    
    for random_state in random_states:
        # Features (X) and target variable (y)
        X = df.drop(target_variable, axis=1)  # Features excluding the target variable
        y = df[target_variable]  # Target variable

        # Convert categorical variables into numerical using one-hot encoding
        X = pd.get_dummies(X, drop_first=True)

        # Split the data into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

        # Standardize the data
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        # Feature selection using SelectKBest with f_regression if specified
        if k_best_features is not None:
            k_best = SelectKBest(f_regression, k=k_best_features)
            X_train_scaled_kbest = k_best.fit_transform(X_train_scaled, y_train)
            X_test_scaled_kbest = k_best.transform(X_test_scaled)

            # Get the names of the selected features
            selected_feature_names = X.columns[k_best.get_support()]
            print(f"Selected Features: {selected_feature_names.tolist()}")

        # Initialize regression models
        models = {
            'Linear Regression': LinearRegression(),
            'Ridge Regression': Ridge(),
            'Lasso Regression': Lasso(),
            'Elastic Net Regression': ElasticNet(),
            'Decision Tree Regressor': DecisionTreeRegressor(random_state=random_state),
            'Random Forest Regressor': RandomForestRegressor(random_state=random_state),
            'Gradient Boosting Regressor': GradientBoostingRegressor(random_state=random_state),
            'XGBoost Regressor': XGBRegressor(),
            'Support Vector Regression': SVR(),
            'Multi-layer Perceptron Regressor': MLPRegressor(random_state=random_state),
            'Bagging Regressor': BaggingRegressor(random_state=random_state),
            'Extra Trees Regressor': ExtraTreesRegressor(random_state=random_state),
            'KNeighbors Regressor': KNeighborsRegressor(),
            'Huber Regressor': HuberRegressor(),
            'SGD Regressor': SGDRegressor(random_state=random_state)
        }

        # Train and evaluate each model
        for model_name, model in models.items():
            model.fit(X_train_scaled_kbest, y_train)
            y_pred = model.predict(X_test_scaled_kbest)

            mse = mean_squared_error(y_test, y_pred)
            r2 = r2_score(y_test, y_pred)

            models_data['model_name'].append(model_name)
            models_data['Random State'].append(random_state)
            models_data['Accuracy'].append(r2)
            models_data['MSE'].append(mse)

            # Save the model to a pickle file
            model_filename = f'output_folder/{model_name.lower().replace(" ", "_")}_model_rs_{random_state}.pkl'
            with open(model_filename, 'wb') as model_file:
                pickle.dump(model, model_file)

    return models_data


target_variable = 'G3'  # Replace with your target variable
random_states = [42, 123, 567, 789, 999]  # Add more random states as needed
results = regression_models_with_feature_selection(df, target_variable, k_best_features=15, random_states=random_states)

# Convert the results to a DataFrame
results_df = pd.DataFrame(results)

results_df
```

## <font color='Brown'>**How does a student’s social life affect the quality of education?**</font>

### **We will be trying to answer with the approach by Chi-square distribution**

### Schoolsup
**Null Hypothesis (H0):** There is no association between the presence of extra educational support ("schoolsup") and the final grades ("G3") of students.

**Alternative Hypothesis (H1):** There is a significant association between the presence of extra educational support ("schoolsup") and the final grades ("G3") of students.

### Famsup
Null Hypothesis (H0): There is no association between family educational support ("famsup") and the final grades ("G3") of students.

Alternative Hypothesis (H1): There is a significant association between family educational support ("famsup") and the final grades ("G3") of students.

### Activities
Null Hypothesis (H0): There is no association between extracurricular activities ("activities") and the final grades ("G3") of students.

Alternative Hypothesis (H1): There is a significant association between extracurricular activities ("activities") and the final grades ("G3") of students.

### Internet
Null Hypothesis (H0): There is no association between internet access at home ("internet") and the final grades ("G3") of students.

Alternative Hypothesis (H1): There is a significant association between internet access at home ("internet") and the final grades ("G3") of students.

### Goout
Null Hypothesis (H0): There is no association between socializing habits ("goout") and the final grades ("G3") of students.

Alternative Hypothesis (H1): There is a significant association between socializing habits ("goout") and the final grades ("G3") of students.

### Romantic
Null Hypothesis (H0): There is no association between romantic relationships ("romantic") and the final grades ("G3") of students.

Alternative Hypothesis (H1): There is a significant association between romantic relationships ("romantic") and the final grades ("G3") of students.

```python
from scipy.stats import chi2_contingency

binary_variables = ['schoolsup', 'famsup', 'activities', 'internet','goout','romantic']

for variable in binary_variables:
    # Create a contingency table
    contingency_table = pd.crosstab(df[variable], df['G3'])
    
    # Print the contingency table
    print(f"Contingency Table for '{variable}' vs. 'G3':")
    print(contingency_table)
    print("\n" + "="*50 + "\n")

    # Perform Chi-squared test
    chi2_value, p_value, _, _ = chi2_contingency(contingency_table)
    print(f"Chi-squared test for '{variable}' vs. 'G3':")
    print(f"Chi-squared value: {chi2_value}")
    print(f"P-value: {p_value}")
    print("\n" + "="*50 + "\n")
```

**'schoolsup' vs. 'G3':**

Interpretation: There is a significant association between receiving extra educational support ('schoolsup') and the final grade ('G3'). Students who receive additional support are more likely to have different final grades compared to those who do not.

**'famsup' vs. 'G3':**

Interpretation: There is no significant association between family educational support ('famsup') and the final grade ('G3'). Family support may not be a determining factor in the variation of final grades.


**'activities' vs. 'G3':**

Interpretation: There is no significant association between participating in extracurricular activities ('activities') and the final grade ('G3'). Engagement in activities outside of academics may not impact final grades.


**'internet' vs. 'G3':**

Interpretation: There is no significant association between internet access at home ('internet') and the final grade ('G3'). The availability of internet access may not be a decisive factor in academic performance.

**'romantic' vs. 'G3':**

Interpretation: There is a significant association between having a romantic relationship ('romantic') and the final grade ('G3'). Students with romantic relationships may experience different academic outcomes compared to those without.

**Meaningful Insights:**

Educational support systems, as indicated by 'schoolsup,' play a significant role in student performance.
The desire for higher education ('higher') strongly correlates with final grades, emphasizing the importance of educational goals.


Family support ('famsup') and participation in extracurricular activities ('activities') do not show significant associations with final grades in this analysis.
Factors such as internet access, attending nursery school, and taking paid classes exhibit non-significant relationships with final grades.
These insights can guide educators and policymakers in tailoring interventions and support programs to address specific factors influencing student performance. Further exploration may involve considering interactions between these variables for a more comprehensive understanding.

## <font color='Brown'>**4. Does Quality of education have a positive or negative relationship with consuming alcohol?**</font>

**Null Hypothesis (H0):**
There is no significant difference in the mean final grades ('G3') between students with different levels of weekday alcohol consumption ('Dalc') and weekend alcohol consumption ('Walc').

**Alternative Hypothesis (H1):**
There is a significant difference in the mean final grades ('G3') between students with different levels of weekday alcohol consumption ('Dalc') and weekend alcohol consumption ('Walc').

```python
from scipy.stats import ttest_ind

# Assuming you have a DataFrame 'df' with columns 'G3', 'Dalc', and 'Walc'
t_statistic_dalc, p_value_dalc = ttest_ind(df[df['Dalc'] > 1]['G3'], df[df['Dalc'] <= 1]['G3'])
t_statistic_walc, p_value_walc = ttest_ind(df[df['Walc'] > 1]['G3'], df[df['Walc'] <= 1]['G3'])

print(f"T-test for Dalc vs. G3: T-statistic={t_statistic_dalc}, P-value={p_value_dalc}")
print(f"T-test for Walc vs. G3: T-statistic={t_statistic_walc}, P-value={p_value_walc}")
```

### Conclusion:
**"T-test for Dalc vs. G3: T-statistic"**

The small p-value indicates strong evidence against the null hypothesis. Therefore, we reject the null hypothesis and conclude that there is a significant difference in mean final grades between students with different levels of weekday alcohol consumption. 

The negative T-statistic suggests that higher weekday alcohol consumption is associated with lower mean final grades.

**"T-test for Walc vs. G3: T-statistic"**

Similarly, the small p-value leads to the rejection of the null hypothesis. This implies a significant difference in mean final grades between students with different levels of weekend alcohol consumption. 

The negative T-statistic suggests that higher weekend alcohol consumption is associated with lower mean final grades.
